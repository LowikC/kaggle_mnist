```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_df = pd.read_csv('../data/train.csv')
```

```python
train_df.head(10)
```

```python
X = np.zeros((train_df.shape[0], 28, 28))
y = np.zeros((train_df.shape[0],))
for i, row in train_df.iterrows():
    label = row.label
    image = np.array(row[1:])
    image = np.asarray(image, dtype=np.float64)
    image = (255.0 - image)/255.0
    image = image.reshape((28, 28))
    X[i,:,:] = image
    y[i] = label
```

```python
def show(X, y, i):
    plt.imshow(X[i,:,:], cmap='gray')
    plt.title('Label: ' + str(int(y[i])))
```

```python
show(X, y, 10)
```
